```python
#given a url http://cnn.com/worldnews/storyoftheday
#this method will strip out the http:// and /storyoftheday
#resulting in cnn.com
def get_domain(url):
    #strip http:// out
    url = re.sub('http\:\/\/', '', url)
    #in the initial run, I forgot to include https
    url = re.sub('https\:\/\/', '', url)

    #strip out trailing URI details so only domain is left
    return re.sub('/.*', '', url)

#url we want a category for
def get_category(domain):
    domain_cat=pd.read_csv(domain_categories_csv)    
    if domain in domain_cat['domain'].values:
        category = domain_cat[domain_cat['domain'] == domain]
        category = category['category'].values
        return category[0]
    else:
        return "Not Categorized"

#check out http://tungwaiyip.info/2012/Collaborative%20Filtering.html
#what we are dong here is just linear correlation
def get_similar_subreddits(subreddit_name,df):
    subreddit = df[subreddit_name]
    return df.corrwith(subreddit)

#takes in a dataframe with subreddits and categories
#returns a dataframe of how similar each subreddit is to each other
def get_subreddit_similarity_matrix(df):
    similarity_matrix = pd.DataFrame(columns=df.columns)

    for subreddit in df.columns:
        s = get_similar_subreddits(subreddit,df)
        similarity_matrix = similarity_matrix.append(s, ignore_index=True)
    return similarity_matrix

#Given a dataframe of posts, and a column
#return a list of series of proportion of each unique value for each subreddit
def get_proportion(df,feature):
    series = []
    subreddit_titles = df['subreddit_title'].unique()

    for subreddit_title in subreddit_titles:

        value_counts = postdf[postdf['subreddit_title'] == subreddit_title][feature].value_counts()
        num_categories = len(postdf[postdf['subreddit_title'] == subreddit_title][feature])


        series.append(pd.Series(value_counts/num_categories, 
                           name=subreddit_title))
    return series

#An implementation of a perumation test
#http://en.wikipedia.org/wiki/Pearson_product-moment_correlation_coefficient
#http://stackoverflow.com/questions/24795535/pythons-implementation-of-permutation-test-with-permutation-number-as-input
def exact_mc_perm_test(xs, ys, nmc):
    n, k = len(xs), 0
    diff = np.abs(np.mean(xs) - np.mean(ys))
    zs = np.concatenate([xs, ys])
    for j in range(nmc):
        np.random.shuffle(zs)
        k += diff < np.abs(np.mean(zs[:n]) - np.mean(zs[n:]))
    return k / nmc
```

```python

data = subreddit_categories.transpose().as_matrix()
#data = subreddit_similarity_matrix.as_matrix()
bandwidth = estimate_bandwidth(data,quantile=0.1)

ms = MeanShift(bandwidth=bandwidth, bin_seeding=True)
ms.fit(data)
labels = ms.labels_
cluster_centers = ms.cluster_centers_

labels_unique = np.unique(labels)
n_clusters_ = len(labels_unique)

print("number of estimated clusters : %d" % n_clusters_)
```
